```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt

try:
    trends = TrendReq(hl='en-US', tz=360, proxies = {'https': 'https://34.203.233.13:80'})
except KeyError as k:
    print(trends)
    
```

```python
trends.build_payload(kw_list=["python"])
data = trends.interest_by_region()
print(data.sort_values(by=['python'],ascending=False).head(15))
```

```python
df = data.sort_values(by=['python'],ascending=False).head(15)
df.reset_index().plot(x="geoName", y="python", figsize=(120,16), kind="bar")
plt.show()
```

```python
data = trends.trending_searches(pn="india")
print(data.head(50))
```

```python
keyword = trends.suggestions(keyword="data science")
data = pd.DataFrame(keyword)
print(data.head(10).to_string())
```
